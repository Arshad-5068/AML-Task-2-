# **Unemployment Analysis with Python**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Unemployment in India.csv")


print("First 5 rows:")
display(df.head())

```

```python
print("\nDataset Info:")
print(df.info())

```

```python
print("\nNull Values:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)

df.columns = [col.strip().replace(" ", "_") for col in df.columns]

print("\nCleaned Columns:")
print(df.columns)
```

```python

print("\nSummary Statistics:")
display(df.describe())


plt.figure(figsize=(10,5))
sns.histplot(df['Estimated_Unemployment_Rate_(%)'], kde=True)
plt.title("Distribution of Unemployment Rate")
plt.xlabel("Unemployment Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='Region', y='Estimated_Unemployment_Rate_(%)', data=df)
plt.xticks(rotation=90)
plt.title("State-wise Unemployment Rate")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(8,5))
numeric_df = df[['Estimated_Unemployment_Rate_(%)', 'Estimated_Employed', 'Estimated_Labour_Participation_Rate_(%)']]
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

    plt.figure(figsize=(12,5))
    sns.lineplot(x='Date', y='Estimated_Unemployment_Rate_(%)', data=df)
    plt.title("Unemployment Trend Over Time")
    plt.xlabel("Date")
    plt.ylabel("Unemployment Rate")
    plt.show()
```
